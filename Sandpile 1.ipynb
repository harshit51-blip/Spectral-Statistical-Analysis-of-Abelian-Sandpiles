```python
list_1 = np.array(np.arange(1,11,1))
list_2 = np.array(np.arange(1,101,1).reshape(10,10))
list_4 = []
list_5 = []
list_6 = []
    
'''b = r.choice(list_2)
b_2 = r.choice(b)'''
    
for t in range(10000):
    for x in list_1:
        a = r.choice(list_1)
        j = np.where(list_1 == a)
        a += 1
        list_1[j] = a
        
        list_3 = [x-1, x+1]
        for o in list_1:
            u = r.uniform(10, 18)
            if o > u:
                b = r.choice(list_3)
                n = np.where(list_1 == o)
                o -= abs(u-o)
                b += abs(u-o)
                list_1[n] = b


                '''if 0 == list_1[0]:
                    list_3 = [x+1]
                    b = r.choice(list_3)
                    n = np.where(list_1 == x)
                    x -= 1
                    b += 1
                    list_1[n] = b'''



                if o == list_1[9]:
                    list_3 = [x-1]
                    b = r.choice(list_3)
                    n = np.where(list_1 == o)
                    o -= abs(u-o)
                    b += abs(u-o)
                    list_1[n] = b
                     
        q = st.mean(list(list_1))
    list_6.append(u)
    list_4.append(q)
    list_5.append(x)
```
