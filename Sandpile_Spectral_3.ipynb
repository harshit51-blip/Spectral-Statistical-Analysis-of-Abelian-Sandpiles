```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import random as r
from itertools import chain
```

```python
#list_1 = np.array(np.arange(1,11,1))
#list_2 = np.array(np.arange(1,101,1).reshape(10,10))
'''b = r.choice(list_2)
b_2 = r.choice(b)'''
def func(N):
    list_5 = []
    list_6 = []
    w_list = []
    list_7 = []
    list_8 = []
    k_list = list(np.arange(3, 15, 1))
    
    
    for t in range(500):
        
        mat = np.random.random((N,N))
        math = mat + mat.conj().T
        
        for x in math:
            a = r.choice(math)
            j_prime = np.where(math == a)
            a_2 = r.choice(a)
            j = np.where(math[j_prime] == a_2)
            a_2 += 1
            math[j_prime][j] = a_2
            list_7.append(a_2)

            list_3 = [x-1, x+1]
            for o in a:
                for k in k_list:
                    u = r.uniform(a_2**k + 5, a_2**(k-1) + 2)
                    if o > u:
                        b = r.choice(list_3)
                        n = np.where(math[j_prime] == o)
                        o -= abs(u-o)
                        b += abs(u-o)
                        math[j_prime][n] = b
                    


                        '''if 0 == list_1[0]:
                            list_3 = [x+1]
                            b = r.choice(list_3)
                            n = np.where(list_1 == x)
                            x -= 1
                            b += 1
                            list_1[n] = b'''



                        if o == math[j_prime][N-1]:
                            list_3 = [x-1]
                            b = r.choice(list_3)
                            n = np.where(math[j_prime] == o)
                            o -= abs(u-o)
                            b += abs(u-o)
                            math[j_prime][n] = b
                               
                list_6.append(u)   
                list_8.append(o)
            list_5.append(x)
        
        
        w = np.linalg.eigvals(mat)
        h = list(w)
        w_list.append(h)
        w_list.sort()
        
        
    return w_list, list_5, list_6, list_7, list_8
```

```python
N_list = [50, 60, 80, 100, 120, 150, 200, 240, 270, 300]
N_2 = [200, 250, 300, 350, 400, 450, 500]
N_3 = list(np.arange(5, 300, 30))
N_4 = [200]
deviation = []
m_list = []
r_dev = []
r_mean = []
c_dev = []
c_mean = []

for i in N_3:
    
    eig_vals, raw_list, crit_points, mean, neigh = func(i)
    raw_list_2 =  list(chain.from_iterable(raw_list))
    eig_vals_2 = list(chain.from_iterable(eig_vals))
    norm_eig = [a/np.sqrt(i) for a in eig_vals_2]
    evd2 = [abs(y - x) for x,y in zip(norm_eig, norm_eig[1:])]
    mean_evd2 = np.mean(evd2)
    rld2 = [y - x for x,y in zip(raw_list_2, raw_list_2[1:])]
    
    
    dev = st.stdev(evd2)
    m = st.mean(evd2)
    dev_r = st.stdev(rld2)
    m_r = st.mean(rld2)
    m_c = st.mean(crit_points)
    dev_c = st.stdev(crit_points)
    deviation.append(dev)
    m_list.append(m)
    r_dev.append(dev_r)
    r_mean.append(m_r)
    c_dev.append(dev_c)
    c_mean.append(m_c)
    
    print('done')
    
    
#mean_evd2 = evd2/np.mean(evd2)
```
