```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import random as r
import seaborn as sns
from itertools import chain
```

```python
#list_1 = np.array(np.arange(1,11,1))
#list_2 = np.array(np.arange(1,101,1).reshape(10,10))
'''b = r.choice(list_2)
b_2 = r.choice(b)'''
def a(g, l, phi, v, d):
    return (g*np.sin(phi)/l) - d*v
def a_w(g, l, phi):
    return (g*np.sin(phi)/l)

def integrator(s):
    deltaT = 0.1
    tFinal = 5.0
    N = int(tFinal/deltaT)

    T = np.zeros(N+1, float)
    xtheta = np.zeros(N+1, float)
    Vx = np.zeros(N+1, float)
    ytheta = np.zeros(N+1, float)
    Vy = np.zeros(N+1, float)


    g = 9.81
    dx = 0.1
    dy = 0.1
    
    lst = list(range(1,101))
    for _ in lst:
        xl = 1.0
        yl = 1.0
        xtheta[0] = r.choice(s)
        Vx[0] = (deltaT/2.0) * a_w(g, xl, xtheta[0])
        ytheta[0] = r.choice(s)
        Vy[0] = (deltaT/2.0) * a_w(g, yl, ytheta[0])

        for n in range(1, N+1):
            T[n] = n * deltaT
            xtheta[n] = xtheta[n-1] + Vx[n-1] * deltaT
            Vx[n] = Vx[n-1] + a(g, xl, xtheta[n], Vx[n-1], dx) * deltaT
            ytheta[n] = ytheta[n-1] + Vy[n-1] * deltaT
            Vy[n] = Vy[n-1] + a(g, yl, ytheta[n], Vy[n-1], dy) * deltaT
            
    return list(xtheta), list(ytheta)





def func(N):
    list_5 = []
    list_6 = []
    w_list = []
    list_7 = []
    list_8 = []
    cout = []
    k_list = list(np.arange(1,10,1))
    counter = 0
    xlist = []
    ylist = []
    
    for t in range(250):
        
        mat = np.random.random((N,N))
        matf = mat + mat.conj().T
        
        
        for i in matf:
            
            a = r.choice(matf)
            j_prime = np.where(matf == a)
            a_2 = r.choice(a)
            j = np.where(matf[j_prime] == a_2)
            a_2 += 1
            matf[j_prime][j] = a_2
            list_7.append(a_2)

            list_3 = [i-1, i+1]
            for o in a:
                for k in k_list:
                    u = r.uniform(a_2**k + 5, a_2**(k-1) + 2)
                    if o > u:
                        b = r.choice(list_3)
                        n = np.where(matf[j_prime] == o)
                        o -= abs(u-o)
                        b += abs(u-o)
                        matf[j_prime][n] = b
                        
            
                    


                        '''if 0 == list_1[0]:
                            list_3 = [x+1]
                            b = r.choice(list_3)
                            n = np.where(list_1 == x)
                            x -= 1
                            b += 1
                            list_1[n] = b'''



                        if o == matf[j_prime][N-1]:
                            list_3 = [x-1]
                            b = r.choice(list_3)
                            n = np.where(matf[j_prime] == o)
                            o -= abs(u-o)
                            b += abs(u-o)
                            matf[j_prime][n] = b
                
                
                   
                list_8.append(o)
                
                list_6.append(u)
            
                
            counter+=1
            
            list_5.append(i)
            cout.append(counter)
        
        
        w = np.linalg.eigvals(mat)
        h = list(w)
        w_list.append(h)
        w_list.sort()
        matf = integrator(a)
        
    return w_list, list_5, list_6, list_7, cout
```

```python
#N_list = [50, 60, 80, 100, 120, 150, 200, 240, 270, 300]
#N_2 = [200, 250, 300, 350, 400, 450, 500]
N_3 = list(np.arange(5, 200, 10))
#N_4 = [120]
list_m = []
list_c = []
deviation = []
r_dev = []
r_mean = []
c_dev = []
c_mean = []
list_mr = []

for p in N_3:
    
    eig_vals, raw_list, crit_points, mean, clist = func(p)
    raw_list_2 =  list(chain.from_iterable(raw_list))
    eig_vals_2 = list(chain.from_iterable(eig_vals))
    norm_eig = [a/np.sqrt(p) for a in eig_vals_2]
    evd2 = [abs(y - x) for x,y in zip(norm_eig, norm_eig[1:])]
    mean_evd2 = evd2/(np.mean(evd2))
    rld2 = [y - x for x,y in zip(raw_list_2, raw_list_2[1:])]
    #d = {'list%s' % i: mean_evd2 for i in range(len(N_3))}
    list_m.append(mean_evd2)
    list_c.append(crit_points)
    dev = st.stdev(evd2)
    m = st.mean(evd2)
    dev_r = st.stdev(rld2)
    m_r = st.mean(rld2)
    m_c = st.mean(crit_points)
    dev_c = st.stdev(crit_points)
    deviation.append(dev)
    list_m.append(m)
    r_dev.append(dev_r)
    r_mean.append(m_r)
    c_dev.append(dev_c)
    c_mean.append(m_c)
    list_mr.append(m_r)
    
    
    print('done')
```
